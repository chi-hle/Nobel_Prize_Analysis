# Nobel Prize Dataset Analysis
The Nobel Prize is possibly the most prestigious scientific honor in the world. Since 1901, the Nobel Prize has recognized extraordinary accomplishments in physics, chemistry, physiology or medicine, literature, and contributions to international peace. Today, it's expanded to one more category which is economics. This dataset includes information about prize winners from 1901 to 2016, let's explore some interesting insights from this dataset!
 

First, the dataset needs to be imported into Python with pandas library

```python
import pandas as pd
import seaborn as sns
nobel_prize= pd.read_csv('/kaggle/input/nobelpeaceprizewinnerproject-dataset/nobel.csv')
nobel_prize.head()
```

```python
nobel_prize.shape
```

```python
nobel_prize.info()
```

As can be seen, there are 911 rows and 18 columns in this datasets. Now, let look into more details of each column

```python
nobel_prize.isnull().sum()
```

There are several columns which contain null values, let's fill these missing data

```python
nobel_prize.fillna(method='ffill')
```

### We have an executive summary about nobel_prize dataset, now let's explore it.

## 0. Who is the prize winner in peace category in 2016

```python
nobel_prize[(nobel_prize['year']==2016) & (nobel_prize['category']=='Peace')]['full_name']
```

```python
nobel_prize.query('year==2016 & category=="Peace"')['full_name']
```

## 1. Which country has the most nobel prize winners?
As can be seen, prize winners came from different nations, but we will question there is any country with dominant winner, is it Russia, America or UK? Let's visualize the prize winners from these five countries from 1901 to 2016.

```python
nobel_prize['birth_country'].value_counts().head(3)
```

=> USA is the country with most nobel prize winners, followings are UK, Germany

## 2. As United States of America has the most nobel prize winners, which is the largest number of USA born given nobel prize in a year?

```python
usa_winner=nobel_prize[nobel_prize['birth_country']=='United States of America']
usa_winner['year'].value_counts().sort_values(ascending=False).head()
```

=> There were 8 USA born people who were given nobel prize in 1972, 1990 and 2001, which is unbelievable! We can look into more details through below visualization about USA born winner proportion from 1901 to 2016.

```python
nobel_prize['usa_born_winner'] = nobel_prize['birth_country']=='United States of America'
nobel_prize['decade'] = ((nobel_prize['year']//10)*10).astype(int)
prop_usa_winners = nobel_prize.groupby('decade',as_index=False)['usa_born_winner'].mean()
line = sns.lineplot(x='decade', y='usa_born_winner', data=prop_usa_winners)
```

## 3. Gender difference between prize winners.
It is usually assumed that in men had achieved more nobel prize than women. Is it true? What is the difference between each gender winner every year?

```python
nobel_prize['sex'].value_counts()
```

=> Sadly, there is huge difference between male and female nobel prize winners. While there are 839 male prize winners, there are only 49 female of them. However, let's look into which category that has the most female winners between 1901 to 2016

```python
female_winner=nobel_prize[nobel_prize['sex']=='Female']
female_winner['category'].value_counts()
```

It is clear that peace is the category that has the most female winners, but how about this figure for male winners?

```python
nobel_prize[(nobel_prize['category']=='Peace')&(nobel_prize['sex']=='Male')]['sex'].count()
```

=> For this category, there are also more male winners than female winners

## 4. Nobel prize winner gender trends in each category during the history

```python
nobel_prize['male_winner'] = nobel_prize['sex']=='Male'
prop_male_winners = nobel_prize.groupby(['decade','category'],as_index=False)['male_winner'].sum()
line = sns.lineplot(x='decade', y='male_winner',hue='category', data=prop_male_winners)
```

```python
nobel_prize['female_winner'] = nobel_prize['sex']=='Female'
prop_female_winners = nobel_prize.groupby(['decade','category'],as_index=False)['female_winner'].sum()
line = sns.lineplot(x='decade', y='female_winner',hue='category', data=prop_female_winners)
```

As can be seen, there was an upward trend in the category of peace for female winners.

## 5. Who is the first woman to get the Nobel prize?

```python
nobel_prize[nobel_prize['sex']=='Female'].sort_values('year').head(1)
```

As can be seen, Marie Curie was the first women to be given the Nobel prize in 1903 in Physics category

## 6. How many nobel prizes were shared between people/organizations?

```python
indiv_winner=nobel_prize['prize_share']== '1/1'
nobel_prize[~indiv_winner]['prize_share'].value_counts()
```

## 7. Who is the youngest and oldest Nobel prize winner?

```python
nobel_prize['birth_date']= pd.to_datetime(nobel_prize['birth_date'])
nobel_prize['age']=nobel_prize['year']-nobel_prize['birth_date'].dt.year
nobel_prize.nsmallest(1, 'age')[['year', 'category','full_name', 'sex','age']]
```

=> The youngest Nobel prize winner got the prize when she was only 17 years old, which is really a huge accomplishment

```python
nobel_prize.nlargest(1, 'age')[['year', 'category','full_name', 'sex','age']]
```

```python
nobel_prize[['year', 'category','full_name', 'sex','age']][nobel_prize['age']==nobel_prize['age'].max()]
```

=> The oldest Nobel prize winner got the prize when he was 90 years old

## 8. What is the most poplular age range of nobel prize winners?

```python
hist=nobel_prize.hist('age')
```

```python
sns.lmplot(x='year', y='age', data=nobel_prize, lowess=True, 
           aspect=2, line_kws={'color' : 'black'})
```

As can be seen, most of the nobel prize winners belong to age range from 53 to 68 years old

## 9. What is average age of prize winners for each category?

```python
nobel_prize.groupby('category')['age'].mean().sort_values()
```

=> Physics is the category with the smallest average age of prize winner while Economics is opposite

## 10. Are there any people who was given Nobel prize more than once?

```python
nobel_prize.groupby('full_name').filter(lambda group: len(group)>=2)['full_name'].unique()
```

There are 6 people/ organizations receiving Nobel prize more than once, they are: Marie Curie, Comit international de la Croix Rouge (International Committee of the Red Cross), Linus Carl Pauling, Office of the United Nations High Commissioner for Refugees (UNHCR), John Bardeen, Frederick Sanger

# 11. Are there any years that Nobel Prize was not organized?

```python
df=pd.DataFrame({'Year':[*range(1901,2017)]})
df[~df['Year'].isin(nobel_prize['year'])]
```
